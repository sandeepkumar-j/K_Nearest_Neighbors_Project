# Import Libraries
Import pandas,seaborn, and the usual libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Get the Data
**Read the 'KNN_Project_Data csv file into a dataframe**

```python
dataframe = pd.read_csv("KNN_Project_Data")
```

**Check the head of the dataframe.**

```python
dataframe.head()
```

# EDA

Since this data is artificial, we'll just do a large pairplot with seaborn.

**Using seaborn on the dataframe to create a pairplot with the hue indicated by the TARGET CLASS column.**

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,7))
sns.pairplot(hue='TARGET CLASS',data=dataframe)
```

# Standardize the Variables

Time to standardize the variables.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(dataframe.drop('TARGET CLASS',axis=1))
```

**Using the .transform() method , to transform the features to a scaled version.**

```python
scaled_features = scaler.transform(dataframe.drop('TARGET CLASS',axis=1))
```

**Convert the scaled features to a dataframe and check the head of this dataframe to make sure the scaling worked.**

```python
df_feat = pd.DataFrame(scaled_features,columns=dataframe.columns[:-1])
df_feat.head()
```

# Train Test Split

**Using train_test_split to split the data into a training set and a testing set.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df_feat,dataframe['TARGET CLASS'],test_size=0.3,random_state=101)
```

# Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

**Creating a KNN model instance with n_neighbors=1**

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

**Fit this KNN model to the training data.**

```python
knn.fit(X_train,y_train)
```

# Predictions and Evaluations

```python
y_pred = knn.predict(X_test)
```

**Create a confusion matrix and classification report.**

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print("Confusion Matrix :\n",confusion_matrix(y_test,y_pred))
```

```python
print("Classification Report :\n",classification_report(y_test,y_pred))
```

# Choosing a K Value

```python
error_rate = []

for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred_i = knn.predict(X_test)
    error_rate.append(np.mean(y_pred_i!=y_test))
```

**Now creating the following plot using the information from your for loop.**

```python
plt.figure(figsize=(10,7))
plt.plot(range(1,40),error_rate,color='blue',linestyle='--',marker='o',markerfacecolor='red',markersize=10)
plt.title('K Value v/s Error rate')
plt.xlabel('K Value')
plt.ylabel('Error Rate')
```

## Retrain with new K Value

```python
# With K value 31
knn = KNeighborsClassifier(n_neighbors=31)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
print("Confusion Matrix and Classification Report with K = 31 ")
print("Confusion Matrix :\n",confusion_matrix(y_test,y_pred))
print("Classification Report :\n",classification_report(y_test,y_pred))
```
